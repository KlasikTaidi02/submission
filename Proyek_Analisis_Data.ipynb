### Cleaning Data

Sintaks dibawah digunakan untuk mengeluarkan kolom yagn tidak relevan dengan hal yang dicari

```python
df_air_quality = df_air_quality.drop(['No','TEMP', 'PRES','DEWP','RAIN','wd', 'WSPM'],axis=1)
```

Data yang kosong akan disis dengan mean dari Data frame. sintaks akan menelusuri setiap kolom di dataframe. Kemudian akan mengisih nilai yang kosong 

```python
list = ['PM2_5','PM10' , 'SO2', 'NO2', 'CO', 'O3']

for i in list:
   df_air_quality[i] = df_air_quality[i].fillna(df_air_quality[i].mean())
   



```

Sintaks dibawah digunakan untuk menngubah nilai outlier menjadi nilai maksimal atau minimum dallam rentar Inter Quantul Range

```python

list = ['PM2_5', 'PM10', 'SO2', 'NO2', 'CO', 'O3']

for column in list:
    Q1 = df_air_quality[column].quantile(0.25)
    Q3 = df_air_quality[column].quantile(0.75)
    IQR = Q3 - Q1

    maximum = Q3 + (1.5*IQR)
    minimum = Q1 - (1.5*IQR)

    more_than = df_air_quality[column] > maximum
    lower_than = df_air_quality[column] < minimum

    df_air_quality[column] = np.where(more_than, maximum, df_air_quality[column])
    df_air_quality[column] = np.where(lower_than, minimum, df_air_quality[column])
    
    
```

## Exploratory Data Analysis (EDA)

### Explore ...

Pada tahap ini, kita melakukan kalkulasi Air Quality Index dengan formula dari CPCB's  AQI Calculator resmi. Pada Kolom - Kolom yanng diperlukan untuk kalkulasi. Kalkukasi akan menghasilkan subindex yang berpengaruh ke AQI. Pada Kolom -Kolom akan dibuat dataframe sendiri untuk menghitung nilai subindexnya. kolom - koloom seperti SO2, NO2 ,PM2_5 dan PM10 akan di aggregat  untuk mendapatkan rata - rata perhari, sedangkan kolom O3 dan CO akan dicari nilai maksimum dari 8 jam

```python
df_average_measure_24hr = df_air_quality.groupby(by=['year','month','day','station']).mean()
 
df_average_measure_24hr = df_average_measure_24hr.drop(['hour','CO', 'O3'], axis=1)

df_average_measure_24hr.reset_index(inplace=True)

df_average_measure_24hr.head()
```

```python
def get_PM2_5_subindex(x):

    if x <= 30:
        return x * 50 / 30
    elif x <= 60:
        return 50 + (x - 30) * 50 / 30
    elif x <= 90:
        return 100 + (x - 60) * 100 / 30
    elif x <= 120:
        return 200 + (x - 90) * 100 / 30
    elif x <= 250:
        return 300 + (x - 120) * 100 / 130
    elif x > 250:
        return 400 + (x - 250) * 100 / 130
    else:
        return 0


df_average_measure_24hr['PM2_5 subindex'] = df_average_measure_24hr['PM2_5'].apply(lambda x: get_PM2_5_subindex(x))

df_average_measure_24hr.head()
```

```python
def get_PM10_subindex(x):
    if x <= 50:
        return x
    elif x <= 100:
        return x
    elif x <= 250:
        return 100 + (x - 100) * 100 / 150
    elif x <= 350:
        return 200 + (x - 250)
    elif x <= 430:
        return 300 + (x - 350) * 100 / 80
    elif x > 430:
        return 400 + (x - 430) * 100 / 80
    else:
        return 0

df_average_measure_24hr['PM10 subindex'] = df_average_measure_24hr['PM10'].apply(lambda x: get_PM10_subindex(x))

df_average_measure_24hr.head()
```

```python
def get_SO2_subindex(x):
    if x <= 40:
        return x * 50 / 40
    elif x <= 80:
        return 50 + (x - 40) * 50 / 40
    elif x <= 380:
        return 100 + (x - 80) * 100 / 300
    elif x <= 800:
        return 200 + (x - 380) * 100 / 420
    elif x <= 1600:
        return 300 + (x - 800) * 100 / 800
    elif x > 1600:
        return 400 + (x - 1600) * 100 / 800
    else:
        return 0

df_average_measure_24hr['SO2 subindex'] = df_average_measure_24hr['SO2'].apply(lambda x: get_SO2_subindex(x))

df_average_measure_24hr.head()  

```

```python
def get_NOx_subindex(x):
    if x <= 40:
        return x * 50 / 40
    elif x <= 80:
        return 50 + (x - 40) * 50 / 40
    elif x <= 180:
        return 100 + (x - 80) * 100 / 100
    elif x <= 280:
        return 200 + (x - 180) * 100 / 100
    elif x <= 400:
        return 300 + (x - 280) * 100 / 120
    elif x > 400:
        return 400 + (x - 400) * 100 / 120
    else:
        return 0
    
df_average_measure_24hr['NO2 subindex'] = df_average_measure_24hr['NO2'].apply(lambda x: get_NOx_subindex(x))

df_average_measure_24hr.head()
```

```python
df_average_measure_8hr_max = df_air_quality.groupby(by=['year','month','day','station']).rolling(window=8, min_periods=1).max().reset_index()
 
df_average_measure_8hr_max = df_average_measure_8hr_max.drop(['level_4','hour','PM2_5','PM10', 'SO2','NO2'], axis=1)

df_average_measure_8hr_max.head(20)
```

Untuk kolom CO, Sub index akan di bagi 10 untuk menjaga keseimbangan data

```python
def get_CO_subindex(x):
    if x <= 1:
        return x * 50 / 1
    elif x <= 2:
        return 50 + (x - 1) * 50 / 1
    elif x <= 10:
        return 100 + (x - 2) * 100 / 8
    elif x <= 17:
        return 200 + (x - 10) * 100 / 7
    elif x <= 34:
        return 300 + (x - 17) * 100 / 17
    elif x > 34:
        return 400 + (x - 34) * 100 / 17
    else:
        return 0

df_average_measure_8hr_max['CO subindex'] = df_average_measure_8hr_max['CO'].apply(lambda x: get_CO_subindex(x)) /10


df_average_measure_8hr_max.head()
```

```python
def get_O3_subindex(x):
    if x <= 50:
        return x * 50 / 50
    elif x <= 100:
        return 50 + (x - 50) * 50 / 50
    elif x <= 168:
        return 100 + (x - 100) * 100 / 68
    elif x <= 208:
        return 200 + (x - 168) * 100 / 40
    elif x <= 748:
        return 300 + (x - 208) * 100 / 539
    elif x > 748:
        return 400 + (x - 400) * 100 / 539
    else:
        return 0
    

df_average_measure_8hr_max['O3 subindex'] = df_average_measure_8hr_max['O3'].apply(lambda x: get_O3_subindex(x))

df_average_measure_8hr_max.head() 



```

```python
df_air_quality_index_AQI_day = pd.merge(df_average_measure_24hr,df_average_measure_8hr_max, on=['year','month', 'day','station'],how='outer')

df_air_quality_index_AQI_day.head()
```

Setelah subIndex didapatkan, maka akan dicari nilai terbesar unutk menwakili AQI

```python
def get_Result(x):
    if x <= 50:
        return "Good"
    elif x <= 100:
        return "Satisfactory"
    elif x <= 200:
        return "Moderate"
    elif x <= 300:
        return "Poor"
    elif x <= 400:
        return "Very Poor"
    elif x > 400:
        return "Severe"
    else:
        return np.NaN
    



df_air_quality_index_AQI_day['AQI Index'] = round(df_air_quality_index_AQI_day[['PM2_5 subindex', 'PM10 subindex', 
                                                                  'SO2 subindex', 'NO2 subindex',
                                                                  'CO subindex', 'O3 subindex']].max(axis=1))

df_air_quality_index_AQI_day['Result'] = df_air_quality_index_AQI_day['AQI Index'].apply(lambda x: get_Result(x))



df_air_quality_index_AQI_day = df_air_quality_index_AQI_day.groupby(by=['station','year','month','day','Result']).mean()

df_air_quality_index_AQI_day.reset_index(inplace=True)


df_air_quality_index_AQI_day.head()



```
